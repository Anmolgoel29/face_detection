```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO("/home/rogue/Desktop/face_recon_media_segregation/models/best.pt")

image = cv2.imread("/home/rogue/Desktop/face_recon_media_segregation/know-faces/anil_mosa_ji.png" , 1)
```

```python
res = model.predict(image)
```

```python
cnt = 0
boxes = res[0].boxes
for i in boxes.xyxy:
    i = list(map(int,i))

    cv2.imwrite(f"image{cnt}.jpg" , image[i[1]:i[3] , i[0]:i[2]])
    cnt+=1

    # cv2.rectangle(image , 
    #               (i[0] , i[1]) , 
    #               (i[2],i[3]) , 
    #               color=(0,255,0),thickness=2)

# cv2.imshow("image",image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input
from keras_facenet import FaceNet
import numpy as np

model = FaceNet()
img = cv2.imread("image0.jpg" , 1)

img = cv2.resize(img , (160,160))
img = image.img_to_array(img)
img = np.expand_dims(img,axis=0)
img = preprocess_input(img)

model.embeddings(img)
```
